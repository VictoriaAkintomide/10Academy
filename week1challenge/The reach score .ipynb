### Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker
```

### Importing the details of influencers

```python
influencers = pd.read_csv('Details_of_influencers.csv')
influencers['reach_score'] = influencers['Followers'] - influencers['Following']
```

### Calculating the reach score of influencers

```python
reach_score_influencers = influencers[['User_Name', 'reach_score', 'Verified', 'Account Created']]
reach_score_influencers = reach_score_influencers.sort_values('reach_score', ascending=False).reset_index(drop=True)
reach_score_influencers
```

### Chart showing the top 10 influencers by reach score

```python
top_10_influencers_reach = reach_score_influencers[:10]
with sns.axes_style("whitegrid"):
    ax = sns.barplot(x='User_Name', y='reach_score', data=top_10_influencers_reach)
ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
ax.yaxis.get_major_formatter().set_scientific(False)
ax.yaxis.get_major_formatter().set_useOffset(False)
plt.xticks(rotation=90)
plt.title('Top 10 african influencers according to their reach score')
plt.xlabel('African Influencers')
plt.ylabel('Reach Score')
plt.savefig('reach_score_influence.png', bbox_inches="tight", dpi=300)
```

### Importing details of government officials

```python
presidents = pd.read_csv('Details_of_presidents.csv')
presidents['reach_score'] = presidents['Followers'] - presidents['Following'] 
```

### Calculating the reach score of government officials

```python
reach_score = presidents[['User_Name', 'reach_score','Verified', 'Account Created']]
reach_score = reach_score.sort_values('reach_score', ascending=False).reset_index(drop=True)
reach_score
```

### Chart showing the top 10 officials by reach score


```python
top_10_presidents_reach = reach_score[:10]
with sns.axes_style("whitegrid"):
    ax = sns.barplot(x='User_Name', y='reach_score', data=top_10_presidents_reach)
ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
ax.yaxis.get_major_formatter().set_scientific(False)
ax.yaxis.get_major_formatter().set_useOffset(False)
plt.xticks(rotation=90)
plt.title('Top 10 african government officials according to their reach score')
plt.xlabel('African government officials')
plt.ylabel('Reach Score')
plt.savefig('reach_score.png', bbox_inches="tight", dpi=300)
```
